```python
import numpy as np
import pandas as pd
from pandas.api.types import is_numeric_dtype

import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('2.4.1 - data.csv')
data
```

  $lnP(10^3/T)$,      

```python
import matplotlib.pyplot as plt

x = np.array(data['10^3/T'])
y = np.array(data['lnP'])

plt.scatter(x, y)

fit = np.polyfit(x, y, 1)
fit_fn = np.poly1d(fit)
plt.plot(x, fit_fn(x))

plt.xlabel('$10^3/T$')
plt.ylabel('$lnP$')

plt.grid()
plt.show()
```

```python
k = fit[0]
k.round(2)
```
