$$C_V = \frac{3}{2}R + \frac{1}{2}2R + \alpha 2R$$
$$C_p = C_V + R$$
   : $$\gamma = \frac{C_p}{C_V}$$

$$=> C_p = C_V \gamma$$
$$C_V = \frac{R}{\gamma -1}$$
$$\frac{R}{\gamma -1} = \frac{3}{2}R + \frac{1}{2}2R + \alpha 2R$$
$$\alpha = \frac{1}{2} (\frac{1}{\gamma - 1} - \frac{5}{2})$$
