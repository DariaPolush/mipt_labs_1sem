```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel(r"/content/электрокапиллярка.xlsx")
```

```python
data.head()
```

```python
data["cos_theta"] = ((data['d']/2)**2-data['h']**2)/((data['d']/2)**2+data['h']**2)
```

```python
data['\sigma'] = 375+51*data['cos_theta']
```

```python
data.head()
```

```python
data1=pd.read_excel(r"/content/электрокапилляркабезадсорбции.xlsx")
```

```python
data1["cos_theta"] = ((data1['d']/2)**2-data1['h']**2)/((data1['d']/2)**2+data1['h']**2)
```

```python
data1['\sigma'] = 375+51*data1['cos_theta']
```

```python
z = np.polyfit(data1['E'], data1['\sigma'], 2)
```

```python
x=np.linspace(-1300, 300, 100)
```

```python
p = np.poly1d(z)
```

```python
z1 = np.polyfit(data['E'], data['\sigma'], 2)
p1 = np.poly1d(z1)
```

```python
plt.scatter(data['E'], data['\sigma'])
plt.xlabel('E, мВ') #Подпись для оси х
plt.ylabel('Surface tension, мН') #Подпись для оси y
plt.title('Электрокапиллярная кривая') #Название
plt.plot(x, p(x), '--', color = 'red', label = 'C учетом адсорбции органических соединений')
plt.plot(x, p1(x), '--', color = 'black', label = 'Без учета адсорбции органических соединений')
plt.grid()
plt.legend()
plt.show()
```
